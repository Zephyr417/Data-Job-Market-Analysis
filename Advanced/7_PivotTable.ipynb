```python
import pandas as pd
import numpy as np
from datasets import load_dataset

# Loading
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df.pivot_table(index='job_title_short',aggfunc='size')
```

```python
df['job_title_short'].value_counts()
```

```python
df.pivot_table(values='salary_year_avg', index='job_title_short',aggfunc='median')
```

```python
df.groupby(by='job_title_short')['salary_year_avg'].agg('median')
```

```python
df.pivot_table(values='salary_year_avg', index='job_country', columns='job_title_short', aggfunc='median')
```

## Top 6 contries median salary for DA,DE,DS

```python
df.head()
```

```python
top_countries = df['job_country'].value_counts().head(6).index
top_countries
```

```python
df_job_country_salary = df.pivot_table(values='salary_year_avg', index='job_country', columns='job_title_short', aggfunc='median')
```

```python
jobs = ['Data Analyst', 'Data Engineer', 'Data Scientist']

df_job_country_salary = df_job_country_salary.loc[top_countries, jobs]
df_job_country_salary
```

```python
df_job_country_salary.plot(kind='bar')

import matplotlib.pyplot as plt

plt.xticks(rotation=30, ha='center', fontsize=10)
plt.xlabel('Job Country')
plt.ylabel('Median of average salary per year')
plt.legend()
plt.show()
```

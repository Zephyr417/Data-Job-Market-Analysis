# Show job posted trends over the year

```python
import pandas as pd
import numpy as np
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
x = [1,2,3,4]
y = [1,2,3,4]
```

```python
plt.plot(x, y)
plt.show()
```

```python
df['job_posted_month'] = df['job_posted_date'].dt.month
```

```python
date_counts = df['job_posted_month'].value_counts()

date_counts = date_counts.sort_index()

date_counts
```

```python
plt.plot(date_counts.index,date_counts)
```

# Bar chart, Counts of job posting

```python
job_counts = df['job_title_short'].value_counts()
job_counts
```

```python
job_counts = job_counts.sort_values(ascending=False)
plt.bar(job_counts.index, job_counts)
plt.title('Postings by Job Title')
plt.ylabel('Number of Postings')
plt.xlabel('Job Titles')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
job_counts = job_counts.sort_values(ascending=True)
plt.barh(job_counts.index, job_counts)
plt.show()
```

```python
job_counts.plot(kind='bar')
plt.title('Postings by Job Title')
plt.ylabel('Number of Postings')
plt.xlabel('Job Titles')
plt.xticks(rotation=45, ha='right')
plt.show()
```

# Exercise: Median Salary by Job Title

```python
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
Median_Salary = df.groupby(by='job_title_short')['salary_year_avg'].median()
Median_Salary = Median_Salary.sort_values(ascending=True)
```

```python
plt.barh(Median_Salary.index, Median_Salary)
plt.title('Median Salary by Job Title')
plt.xlabel('Job Title')
plt.ylabel('Median Salary')
plt.show()
```
